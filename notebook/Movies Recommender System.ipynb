```python
# it is the list of dict
# we have to convert [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# to [Action,Adventure,Fantasy,Science Fiction]

```

```python
import ast
#  convert a string that looks like a list or dictionary into an actual list or dictionary.
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# same for keywords
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(2)
```

```python
# fetching top 3 actors
```

```python
def convert3(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt != 3:
            L.append(i['name'])
            cnt += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

```python
# fethcing the director
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
type(movies['overview'][0]) # -> it is str we will convert it into list so we can concatinate 
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
# we will remove space between words Sam Worthington -> SamWorthington
```

```python
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# we will make tags column by concatinating overview, overview, keywords, cast, crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# converting tags(list) to str
new_df['tags'] = new_df['tags'].apply(lambda x :" " .join(x))
```

```python
import nltk 
```

```python
# for removing similer word from max_features during vectorization
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
# Ex: 
print(ps.stem('loved'))
print(ps.stem('loving'))
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
# converting to lower case letters
new_df['tags'] = new_df['tags'].apply(lambda x :x.lower())
```

```python
new_df.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# 5 similar movies to movie at index 
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
